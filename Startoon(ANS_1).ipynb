```python
import numpy as np
import matplotlib.pyplot as plt

data_1 = np.loadtxt('Data_1.txt')
data_2 = np.loadtxt('Data_2.txt')

def find_peaks(data):
    maxima = []
    minima = []

    for i in range(1, len(data) - 1):
        if data[i] > data[i - 1] and data[i] > data[i + 1]:
            maxima.append(i)
        elif data[i] < data[i - 1] and data[i] < data[i + 1]:
            minima.append(i)

    return maxima, minima

maxima_1, minima_1 = find_peaks(data_1)
maxima_2, minima_2 = find_peaks(data_2)


def plot_data_with_peaks(data, maxima, minima, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data, label='Signal')
    plt.scatter(maxima, data[maxima], color='red', label='Maxima')
    plt.scatter(minima, data[minima], color='blue', label='Minima')
    plt.title(title)
    plt.xlabel('Index')
    plt.ylabel('Signal Value')
    plt.legend()
    plt.show()

plot_data_with_peaks(data_1, maxima_1, minima_1, 'Data Set 1')
plot_data_with_peaks(data_2, maxima_2, minima_2, 'Data Set 2')

# Print indices of maxima and minima
print(f"Data Set 1 Maxima Indices: {maxima_1}")
print(f"Data Set 1 Minima Indices: {minima_1}")
print(f"Data Set 2 Maxima Indices: {maxima_2}")
print(f"Data Set 2 Minima Indices: {minima_2}")
```
